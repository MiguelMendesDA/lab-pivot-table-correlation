### Bonus

#### We now want to compute the mean salary after removing the maxima for each Department. Create the appropriate pivot table.

Hint: Write a custom function.

```python
# Custom function to calculate mean salary after removing maxima
def mean_salary_after_removing_max(series):
    max_value = series.max()
    series = series[series < max_value]  # Remove the max value
    if series.empty:
        return 0  # Return 0 if all values are removed
    return series.mean()

# Create a pivot table using the custom function
pivot_table = pd.pivot_table(employee, values='Salary', index=['Department', 'Title'], aggfunc=mean_salary_after_removing_max)

# Display the pivot table
print(pivot_table)
```

#### Is there a correlation between study time and Grade? Use both Pearson and Spearman correlation and comment your results.

```python
# your answer here
import pandas as pd
from scipy.stats import pearsonr, spearmanr

# Assuming df is your DataFrame with the 'study_time' and 'grade' columns
# Replace df with your actual DataFrame


# Calculate Pearson correlation
pearson_corr, _ = pearsonr(time_grades['Study time'], time_grades['Grade'])
print("Pearson Correlation:", pearson_corr)

# Calculate Spearman correlation
spearman_corr, _ = spearmanr(time_grades['Study time'], time_grades['Grade'])
print("Spearman Correlation:", spearman_corr)

# Comment on the results
if abs(pearson_corr) > abs(spearman_corr):
    stronger_correlation = 'Pearson'
else:
    stronger_correlation = 'Spearman'

print(f"\nThe correlation between study time and grade is stronger using {stronger_correlation} correlation.")

# Display scatter plot
plt.scatter(time_grades['Study time'], time_grades['Grade'])
plt.title('Scatter Plot of Study Time vs Grade')
plt.xlabel('Study Time')
plt.ylabel('Grade')
plt.show()
```

#### Could you explain the difference between Pearson and Spearman correlation? (see References in README.md) 

# your answer here
Pearson Correlation:

Type of Relationship:

Measures the linear relationship between two continuous variables.
Assumes a linear relationship, meaning that a change in one variable is associated with a proportional change in the other.

Applicability:

Suitable for variables with a linear relationship.
Sensitive to outliers and may be affected by them.


Calculation:

Ranges from -1 to 1.
Positive values indicate a positive linear relationship.
Negative values indicate a negative linear relationship.
0 indicates no linear relationship.


Spearman Correlation:


Type of Relationship:

Measures the monotonic relationship between two variables.
Monotonicity means that as one variable increases, the other tends to also increase or decrease, but not necessarily at a constant rate.


Applicability:

Suitable for variables with a non-linear relationship.
Less sensitive to outliers.


Calculation:

Ranges from -1 to 1.
Similar interpretation to Pearson correlation in terms of direction (positive, negative, or no correlation).
Can capture relationships that Pearson correlation might miss if the relationship is not strictly linear.


Summary:

Both correlations range from -1 to 1, where 1 indicates a perfect positive correlation, -1 indicates a perfect negative correlation, and 0 indicates no correlation.

Pearson correlation is suitable for linear relationships and assumes normal distributions.
Spearman correlation is suitable for detecting monotonic relationships and is less affected by outliers.

In summary, the choice between Pearson and Spearman correlation depends on the nature of the relationship in your data and the assumptions you can make about the data distribution. If the relationship is linear, Pearson correlation may be appropriate, while Spearman correlation is more robust to non-linear relationships and outliers.
